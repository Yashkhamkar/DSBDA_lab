```python
import bs4 
import requests
```

```python
request1 = requests.get('https://www.flipkart.com/nike-air-zoom-vomero-16-running-shoes-men/p/itmc5375def6449e?pid=SHOG3GB3DHSFMZUP&lid=LSTSHOG3GB3DHSFMZUP7B3WHQ&marketplace=FLIPKART&q=nike+shoes&store=osp&srno=s_1_6&otracker=search&otracker1=search&fm=Search&iid=0e3733c6-7873-43cd-b286-7764e96b0f10.SHOG3GB3DHSFMZUP.SEARCH&ppt=sp&ppn=sp&ssid=ps3lr05exyzs7uv41684647412799&qH=2d7d99166bc4a50f')
```

```python
request1
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

```python
soup
```

```python
review = soup.findAll('div',{'class':'_6K-7Co'})
```

```python
for reviews in review:
    print(reviews.get_text()+'\n\n')
```

```python
#Overall Avg. rating

rating = soup.find('div',{'class':'_3LWZlK _138NNC'}).get_text()
rating
```

```python
#Individual rating
ind_rating = soup.findAll('div',{'class':'_3LWZlK _1rdVr6 _1BLPMq _3B8WaH'})
for individual in ind_rating:
    print(individual.get_text() + '\n\n')
    
```

```python
#Customer Name

cust_name = soup.findAll('p',{'class':'_2mcZGG'})
for customer in cust_name:
    print(customer.get_text() + '\n\n')
```

```python
#prodcut name
name = soup.find('span',{'class':'B_NuCI'}).get_text()
```
